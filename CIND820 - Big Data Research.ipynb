```python
import numpy as np
import pandas as pd
```

```python
googleApps = pd.read_csv("/Users/user/Documents/CIND820 - Big Data Analytics Project/Google Play Store Apps/googleplaystore.csv", delimiter = ',')
googleApps.info()
```

```python
###dropping NA values
googleApps.dropna(inplace = True)
```

```python
##Converting Application Name to Category Type
googleApps["Application Name"] = googleApps["Application Name"].astype('category')
```

```python
#Converting Rating to string 
googleApps["Rating"] = googleApps["Rating"].astype(str)
```

```python
def convert_rating(rating):
    if rating[0] == "1":
        rating = "Low"
        return rating
    elif rating[0] == "2":
        rating = "Low"
        return rating
    elif rating[0] == "3":
        rating = "Medium"
        return rating
    elif rating[0] == "4":
        rating = "High"
        return rating
    elif rating[0] == "5":
        rating = "High"
        return rating
```

```python
googleAppRating = map(convert_rating,googleApps['Rating'])
googleApps['Rating'] = list(googleAppRating)
```

```python
##Converting Category to Category Type
googleApps["Category"] = googleApps["Category"].astype('category')
```

```python
###Converting Reviews to Numerical
googleApps['Reviews'] = pd.to_numeric(googleApps['Reviews'])
```

```python
##Converting Size to Numeric, remove period 
def convert_size(size):
    if 'M' in size:
        num = size[:-1]
        num = float(num) * 1000000
        return num
    elif 'k' in size:
        num = size[:-1] 
        num = float(num) * 1000
        return num 
```

```python
googleAppsSize = map(convert_size, googleApps['Size'])
googleApps['Size'] = list(googleAppsSize)
```

```python
### Converting Installs to Numerical and removing "+" and ','
googleApps['Installs'] = googleApps['Installs'].str.replace(',', '')
googleApps['Installs'] = googleApps['Installs'].str.replace('+', '')
googleApps['Installs'] = googleApps['Installs'].replace('Free', np.NaN)
googleApps['Installs'] = pd.to_numeric(googleApps['Installs'])  
```

```python
print((googleApps)['Installs'].head())
```

```python
###Changing Type Variable into dummy variable
googleApps['Type'] = pd.get_dummies(googleApps['Type'])
print((googleApps)['Type'].head())    ### Free = 1 Paid = 0
```

```python
###Changing Price into Numerical
def convert_price(price):
    if "$" in  price:
        x = price.replace('$','')
        price = float(x)
        return price
    elif 'Everyone' == price:
        return None
    else:
        x = float(price)
        return x   
```

```python
googleAppPrice = map(convert_price,googleApps['Price'])
googleApps['Price'] = list(googleAppPrice)
googleApps['Price'].head()
```

```python
### Convert Content Rating into dummy variables
#googleApps["Content Rating"] = googleApps["Content Rating"].astype('category')
#googleApps["numericContentRating"] = googleApps["Content Rating"].cat.codes

def numeric_content_rating(rating):
    if rating == "Adults only 18+":
        return int(1)
    elif rating == "Everyone":
        return int(2)
    elif rating == "Everyone 10+":
        return int(3)
    elif rating == "Mature 17+":
        return int(4)
    elif rating == "Teen":
        return int(5)
    else:
        return int(6)
```

```python
googleAppContentRating = map(numeric_content_rating,googleApps['Content Rating'])
googleApps['Content Rating'] = list(googleAppContentRating)
```

```python
## Convert Genres to Categorical
googleApps['Genres'] = googleApps["Genres"].astype('category')
```

```python
def single_genre(genre):
    genre = genre.split(';')
    genre = genre[0]
    return genre
```

```python
googleAppGenre = map(single_genre,googleApps['Genres'])
googleApps['Genres'] = list(googleAppGenre)
```

```python
##Get all the unique genre 

def get_genres(genres):
    uniqueList = []
    for genre in  genres:
        genre = genre.split(';')
        if genre not in uniqueList:
            uniqueList.append(genre)
    return uniqueList

result = get_genre(googleApps['Genres'])

def unique_genres(result):
    uniqueGenre = []
    for genre in result:
        if genre[0] not in uniqueGenre:
            uniqueGenre.append(genre[0])
    return uniqueGenre

result2 = unique_genres(result)
```
